```
# funkcia na ohodnotenie modelu
def evaluate(model, X_test, y_test):
    y_pred=model.predict(X_test)
    y_pred_p=model.predict_proba(X_test)
    class_report = classification_report(y_test.ravel(),y_pred)
    conf_matrix = metrics.confusion_matrix(y_test, y_pred)
    model_score = model.score(X_test, y_test)
    roc_auc = metrics.roc_auc_score(y_test, y_pred_p[:,1]) #toto mozem potom zmazat ked model score bude roc_auc
  
    print(class_report)
    print(conf_matrix)
    print('Model score is: {:0.3f}'.format(model_score)) 
    print('ROC_AUC is: {:0.3f}'.format(roc_auc))

# trenovacie a validacne metriky
def cv_score(grid_result):
  # ROC_AUC
  means_test_roc_auc = grid_result.cv_results_['mean_test_roc_auc']
  means_train_roc_auc = grid_result.cv_results_['mean_train_roc_auc']
  params = grid_result.cv_results_['params']

  for mean_test, mean_train, param in zip(means_test_roc_auc, means_train_roc_auc, params):
    if param == grid_result.best_params_:
      print('ROC_AUC valid (train) %f (%f) with: %r' % (mean_test, mean_train, param))

# RECALL
  means_test_recall = grid_result.cv_results_['mean_test_recall']
  means_train_recall = grid_result.cv_results_['mean_train_recall']
  params = grid_result.cv_results_['params']

  for mean_test, mean_train, param in zip(means_test_recall, means_train_recall, params):
    if param == grid_result.best_params_:
      print('RECALL valid (train) %f (%f) with: %r' % (mean_test, mean_train, param))

# PRECISION
  means_test_precision = grid_result.cv_results_['mean_test_precision']
  means_train_precision = grid_result.cv_results_['mean_train_precision']
  params = grid_result.cv_results_['params']

  for mean_test, mean_train, param in zip(means_test_precision, means_train_precision, params):
    if param == grid_result.best_params_:
      print('PRECISION valid (train) %f (%f) with: %r' % (mean_test, mean_train, param))   
```

```
# RANDOM FOREST, RandomizedSearchCV

pipe = Pipeline(steps=[('classifier', RandomForestClassifier(random_state=1, class_weight='balanced'))]) 

n_estimators = [100, 500]
max_depth = [2, 4]
min_samples_split = [5, 25, 35]
min_samples_leaf = [10, 20, 30]

random_grid = {'classifier__n_estimators': n_estimators,
               'classifier__max_depth': max_depth,
               'classifier__min_samples_split': min_samples_split,
               'classifier__min_samples_leaf': min_samples_leaf}

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=3) 

grid = RandomizedSearchCV(pipe, random_grid, n_iter=20, cv=cv, scoring=['recall', 'roc_auc', 'precision'], refit='roc_auc', return_train_score=True, n_jobs=-1, verbose=1, random_state = 45) 

grid_result = grid.fit(X_train, y_train.ravel()) 

best_model = grid_result.best_estimator_
```

```
def plot_results(model, param = 'n_estimators', name = 'Num Trees'):
    param_name = 'param_classifier__n_estimators' #'param_%s' % param

    # Extract information from the cross validation model
    train_scores = model.cv_results_['mean_train_roc_auc']
    test_scores = model.cv_results_['mean_test_roc_auc']
    train_time = model.cv_results_['mean_fit_time']
    param_values = [1, 13, 25, 37, 49, 62, 74, 86, 98, 111, 123, 135, 147,
                    159, 172, 184, 196, 208, 221, 233, 245, 257, 269, 282,
                    294, 306, 318, 331, 343, 355, 367, 379, 392, 404, 416,
                    428, 441, 453, 465, 477, 489, 502, 514, 526, 538, 551,
                    563, 575, 587, 600] #list(model.cv_results_[param_name])
    
    # Plot the scores over the parameter
    plt.subplots(1, 2, figsize=(10, 6))
    plt.subplot(121)
    plt.plot(param_values, train_scores, 'bo-', label = 'train')
    plt.plot(param_values, test_scores, 'go-', label = 'test')
    plt.ylim(ymin = -10, ymax = 0)
    plt.legend()
    plt.xlabel(name)
    plt.ylabel('Neg Mean Absolute Error')
    plt.title('Score vs %s' % name)
    
    plt.subplot(122)
    plt.plot(param_values, train_time, 'ro-')
    plt.ylim(ymin = 0.0, ymax = 2.0)
    plt.xlabel(name)
    plt.ylabel('Train Time (sec)')
    plt.title('Training Time vs %s' % name)
    
    
    plt.tight_layout(pad = 4)
```

```
import matplotlib.pyplot as plt
import numpy as np

def gini(p):
   return (p)*(1 - (p)) + (1 - p)*(1 - (1-p))

def entropy(p):
   return - p*np.log2(p) - (1 - p)*np.log2((1 - p))

def classification_error(p):
   return 1 - np.max([p, 1 - p])

x = np.arange(0.0, 1.0, 0.01)
ent = [entropy(p) if p != 0 else None for p in x]
scaled_ent = [e*0.5 if e else None for e in ent]
c_err = [classification_error(i) for i in x]

fig = plt.figure()
ax = plt.subplot(111)

for j, lab, ls, c, in zip(
      [ent, scaled_ent, gini(x), c_err],
      #['Entropy', 'Entropy (scaled)', 'Gini Impurity', 'Misclassification Error'],
      ['Entropia', 'Gini Index'],
      ['-', '-', '--', '-.'],
      ['lightgray', 'red', 'green', 'blue']):
   line = ax.plot(x, j, label=lab, linestyle=ls, lw=1, color=c)

ax.legend(loc='upper left', bbox_to_anchor=(0.01, 0.9),
         ncol=1, fancybox=True, shadow=False)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')

plt.ylim([0, 1.1])
plt.xlabel('p^(mk)') #ale len p^ (plati pre konkretne 'm' a 'k')
#plt.ylabel('Kritérium') # ako inak to nazvat? metrika? nakladova funkcia? kritérium
plt.show()
```
